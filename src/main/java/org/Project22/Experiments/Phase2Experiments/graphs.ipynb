```python
import numpy as np
import matplotlib.pyplot as plt
```

### Experiment 1 - Loading the CFG

```python
# load all data into vectors
cyk_load_small = np.loadtxt("cyk_load_cfg_small.txt") / 1000000
cyk_load_medium = np.loadtxt("cyk_load_cfg_medium.txt") / 1000000
cyk_load_large = np.loadtxt("cyk_load_cfg_large.txt") / 1000000
tt_load_small = np.loadtxt("tree_traversal_load_cfg_small.txt") / 1000000
tt_load_medium = np.loadtxt("tree_traversal_load_cfg_medium.txt") / 1000000
tt_load_large = np.loadtxt("tree_traversal_load_cfg_large.txt") / 1000000
```

```python
x = ['Small', 'Medium', 'Large']
y_cyk = [cyk_load_small.mean(), cyk_load_medium.mean(), cyk_load_large.mean()]
y_tt = [tt_load_small.mean(), tt_load_medium.mean(), tt_load_large.mean()]

x_axis = np.arange(len(x))

plt.bar(x_axis - 0.2, y_cyk, 0.4, color='darkorange', edgecolor='black', label='CYK')
plt.bar(x_axis + 0.2, y_tt, 0.4, color='royalblue', edgecolor='black', label='Tree Traversal')
  
plt.xticks(x_axis, x)
plt.xlabel("CFG Size")
plt.ylabel("Avg. Load Time (ms)")
plt.title("Load Time per CFG Size")
plt.legend()
plt.show()
```
